## 물놀이 장소 데이터 전처리

```python
# 물놀이 장소 데이터
WP = pd.read_csv('../../raw_data/물놀이/물놀이_스마트서울맵_geojson.csv', encoding='euc-kr')
wp = WP.copy()
wp.head()
```

```python
# 사용할 컬럼만 남기기
wp = wp[['CONTENTS_NAME', 'IMG_MAIN_URL', 'ADDR_OLD', 'ADDR_NEW', 'COORD_DATA', 'TEL_NO', 'EXTRA_DATA_02', 
              'NAME_01', 'VALUE_01', 'NAME_02', 'VALUE_02', 'NAME_03', 'VALUE_03', 'NAME_04', 'VALUE_04', 
              'NAME_05', 'VALUE_05', 'NAME_06', 'VALUE_06', 'NAME_07', 'VALUE_07', 'SUB_ID', 'SUB_NAME']]
```

```python
# 수영장 - 한강 야외 수영장 4군데만 남기기

temp_list = ['광나루 수영장', '여의도 수영장', '잠원 수영장', '뚝섬 수영장']
empty_list = []

temp_df = wp[wp.SUB_NAME == '수영장']

# temp_df

for i in range(len(temp_df)):
    if temp_df.iloc[i].CONTENTS_NAME in temp_list:
        empty_list.append(temp_df.iloc[i])
        
temp_df = pd.DataFrame(empty_list)
temp_df
```

```python
# 바닥분수, 물놀이장, 한강야외수영장 -> 총 140개 남기기

wp = wp[(wp.SUB_NAME == '바닥분수') | (wp.SUB_NAME == '물놀이장')]

wp = pd.concat([wp, temp_df], axis=0)

wp.head()
```

### 위도, 경도 분리 후 파생변수 따로 생성

```python
la, lo = [], []

for i in wp.COORD_DATA:
    i = i.replace('[', '')
    i = i.replace(']', '')
    x, y = i.split(',')
    lo.append(x)
    la.append(y)
    
wp['LA'] = la
wp['LO'] = lo

wp.drop(columns=['COORD_DATA'], axis=1, inplace=True)

wp.head()
```

## 물놀이_서울시 소방서,안전센터,구조대 위치정보 데이터 전처리

```python
# 안전시설_경찰및소방 데이터
SAFE119 = pd.read_csv('../../raw_data/물놀이/물놀이_서울시 소방서,안전센터,구조대 위치정보.csv', encoding='euc-kr')
safe119 = SAFE119.copy()
safe119.drop(['연번', '도서지역포함여부', '상위서ㆍ센터ID', '일련번호'], axis=1, inplace=True)
safe119.head()
```

```python
# 좌표 변환
# ITRF2000_MTM -> WGS84
import numpy as np
import pyproj
import folium

def project_array(coord, p1_type, p2_type):
    """
    좌표계 변환 함수
    - coord: x, y 좌표 정보가 담긴 NumPy Array
    - p1_type: 입력 좌표계 정보 ex) epsg:5179
    - p2_type: 출력 좌표계 정보 ex) epsg:4326
    """
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]
```

```python
# numpy 객체로 변환
safe119_num = np.array(safe119[['FCLTY_LO', 'FCLTY_LA']])
safe119_num
```

```python
# 좌표계 정보 설정
p1_type = "epsg:5186"
p2_type = "epsg:4326"

# project_array() 함수 실행
result = project_array(safe119_num, p1_type, p2_type)
result
```

```python
safe119['FCLTY_LO'] = result[:, 0]
safe119['FCLTY_LA'] = result[:, 1]
safe119
```

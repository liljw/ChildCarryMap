## 2. 팝업창에 쓰일 value01 저장

```python
def get_shortvalue(s):
    return s.split('\n')[0]

wp['waterplayshortvalue01'] = wp['waterplayvalue01'].apply(get_shortvalue)
```

```python
wp['waterplayshortvalue01']
```

## 3. 상세페이지에 쓰일 해당 자치구의 평균 점수

```python
def get_signgu(s):
    return s.split(' ')[1]

# 각 물놀이장의 행정구역 저장 
wp['waterplayscoremean'] = wp['waterplayaddrold'].apply(get_signgu)
```

```python
# 행정구역 별 물놀이장의 점수 평균 
waterplayscoremean = wp['waterplayscorefaci'].groupby(wp['waterplayscoremean']).mean()
waterplayscoremean
```

```python
# 각 물놀이장이 속한 행정구역의 점수 평균 저장 
for i in range(len(wp)):
    mysigngu = wp['waterplayscoremean'][i]
    
    wp['waterplayscoremean'][i] = waterplayscoremean[mysigngu]
```

```python
wp['waterplayscoremean']
```

## 7. 방사형 차트에 쓰일 물놀이장 종류에 따른 편의시설 별 점수 평균 값 삽입

```python
# 물놀이장 종류에 따른 편의시설 별 편의지수 
wpfaciscoremean = wp.groupby('waterplaytypeid')['waterplayscoretoilet', 'waterplayscoreconv', 'waterplayscoredrug',
                                                'waterplayscoremedi', 'waterplayscoresafe112', 'waterplayscoresafe119',
                                                'waterplayscoreparking',
                                               ].mean()
```

```python
wpfaciscoremean.columns = ['waterplayscoretoiletmean', 'waterplayscoreconvmean', 'waterplayscoredrugmean',
                           'waterplayscoremedimean', 'waterplayscoresafe112mean', 'waterplayscoresafe119mean',
                           'waterplayscoreparkingmean']
wpfaciscoremean
```

```python
# 임시 df 생성 
_l = [0 for i in range(7)]
_l
_df = pd.DataFrame([_l for i in range(len(wp))])
_df.columns = ['waterplayscoretoiletmean', 'waterplayscoreconvmean', 'waterplayscoredrugmean', 'waterplayscoremedimean',
               'waterplayscoresafe112mean', 'waterplayscoresafe119mean', 'waterplayscoreparkingmean']
_df
```

```python
# 임시 df에 물놀이장 종류에 따른 편의시설 별 편의지수 삽입 
for i in range(len(wp)):
    if wp.iloc[i]['waterplaytypeid'] == 1:
        _df.iloc[i] = wpfaciscoremean.iloc[0]
    elif wp.iloc[i]['waterplaytypeid'] == 2:
        _df.iloc[i] = wpfaciscoremean.iloc[1]
    else:
        _df.iloc[i] = wpfaciscoremean.iloc[2]

_df
```

```python
# 임시 df를 wp에 추가 
wp = pd.concat([wp, _df], axis=1)
wp
```

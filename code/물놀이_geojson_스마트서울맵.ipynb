```python
import pandas as pd
import json
```

```python
geo_dir = '../../raw_data/100765_20230714225635.geojson'
geo_json = json.load(open(geo_dir, encoding='utf-8'))
```

```python
geo_json
```

```python
geo_feature_list = geo_json.get('features')
geo_feature_list
```

```python
type_list = []
geometry_type_list = []
geometries_type_list = []

for i in range(len(geo_feature_list)):
    type_list.append(geo_feature_list[i].get('type'))
    geometry_type_list.append(geo_feature_list[i].get('geometry').get('type'))
    geometries_type_list.append(geo_feature_list[i].get('geometry').get('geometries')[0].get('type'))
```

```python
geometry_point_df = pd.DataFrame()
properties_df = pd.DataFrame()
polygon_count = 0

for i in range(len(geo_feature_list)):
    _df1 = pd.DataFrame(geo_feature_list[i].get('geometry').get('geometries')[0].get('coordinates')).transpose()
    # 그늘막 텐트의 경우 폴리곤 좌표이므로 생략
    if len(_df1) != 1:
        polygon_count += 1
        continue
    geometry_point_df = pd.concat([geometry_point_df, _df1])
    
    _df2 = pd.DataFrame([geo_feature_list[i].get('properties')])
    properties_df = pd.concat([properties_df, _df2])
    
geometry_point_df.columns = ['x', 'y']
```

```python
from collections import Counter

Counter(type_list)
Counter(geometry_type_list)
Counter(geometries_type_list)
polygon_count
```

```python
result_df = pd.concat([geometry_point_df, properties_df], axis=1)
```

```python
result_df
```

```python
result_df.to_csv('../../raw_data/물놀이_스마트서울맵_geojson.csv', index=False, encoding='EUC-KR')
```
